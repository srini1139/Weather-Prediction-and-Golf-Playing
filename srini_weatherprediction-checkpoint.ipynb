```python
#Replace the string as numeric
df['Windy'] = df['Windy'].astype('str')
obj_numbers = {'Outlook': {'sunny':1,'overcast':2,'rainy':3},
              'Temperature' :{'hot':1,'cool':2,'mild':3},
              'Humidity': {'high':1,'normal':2},
              'Windy': {'False':0, 'True':1},
              'Play':{'no':0,'yes':1}}



df.replace(obj_numbers,inplace = True)
```

```python
#convert all the columns as category
columns = []
for i in df.columns:
    columns.append(i)
print (columns)
for i in columns:
 df[i] = df[i].astype('category')
df.info()
```

```python
#perform distribution plot to see the distribution of the attributes
import itertools
columns = [c for c in df.columns if c not in 'Play']
colours = ["b","r","g","gold"]
length = len(columns)
fig = plt.figure(figsize=(20,20))

for i,j,k in itertools.zip_longest(columns,range(length),colours):
    plt.subplot(4,3,j+1)
    ax = sns.distplot(df[i],bins = 10, color = k)
    ax.set_facecolor("w")
    plt.legend(loc="best")
    plt.title(i,color="black")
    plt.xlabel("")
```

# Split the data into train and test

```python
X = df.drop(columns = 'Play',axis = 1)
y = df[['Play']]
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,  test_size=0.25, random_state=20)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

# Checking the class distribution in the train set

```python
a1 = len(y_train[y_train['Play'] != 'no' ])
a1
a2 = len(df) - a1
a2
a3 = [a2,a1]
labels = ['Not Playing Golf','Play Golf']
colors = ['g','r']
plt.pie(a3,labels = labels,colors = colors,autopct='%.1f%%')
plt.title('Golf Label split')
plt.show()
```

# Decision Tree Model

```python
model = DecisionTreeClassifier(ccp_alpha=0.0, class_weight=None, criterion='entropy',
                       max_depth=None, max_features=None, max_leaf_nodes=None,
                       min_impurity_decrease=0.0, min_impurity_split=None,
                       min_samples_leaf=1, min_samples_split=2,
                       min_weight_fraction_leaf=0.0, presort='deprecated',
                       random_state=None, splitter='best' )
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)
print ('Train accuracy %f' % accuracy_score(y_train, y_train_pred))
print ('Test accuracy: %f' % accuracy_score(y_test, y_pred))
print ('recall %f' % recall_score(y_test, y_pred, average='weighted'))
print ('precision %f' % precision_score(y_test, y_pred, average='weighted'))
```

```python
print (confusion_matrix(y_test, y_pred))
print("classification Report")
print(metrics.classification_report(y_test, y_pred))
resultsDf = pd.DataFrame({'Method':['Decision Tree'], 'Train Score': (accuracy_score(y_train, y_train_pred))*100.0,
                                  'Test Score': (accuracy_score(y_test, y_pred))*100.0})
resultsDf = resultsDf[['Method', 'Train Score', 'Test Score']]    
resultsDf
```

```python
from imblearn.over_sampling import SMOTE
```

```python
print("Before UpSampling, counts of label '1': {}".format(len(y_train[y_train['Play'] == 1 ])))
print("Before UpSampling, counts of label '0': {} \n".format(len(y_train[y_train['Play'] == 0 ])))

sm = SMOTE(sampling_strategy = 1 ,k_neighbors = 2, random_state=1)   #Synthetic Minority Over Sampling Technique
X_train, y_train = sm.fit_sample(X_train, y_train)


print("After UpSampling, counts of label '1': {}".format(len(y_train[y_train['Play'] == 1 ])))
print("After UpSampling, counts of label '0': {} \n".format(len(y_train[y_train['Play'] == 0 ])))



print('After UpSampling, the shape of train_X: {}'.format(X_train.shape))
print('After UpSampling, the shape of train_y: {} \n'.format(y_train.shape))



```
